```python
import cv2
import matplotlib.pyplot as plt
```

```python
img_path = '../annotated_frames/data_2/10_webcam_2.jpg'
object_bbox_txt_path = '../objects_bounding_boxes/data_1/10_objects_bbox_2.txt'

img = cv2.imread(img_path)
with open(object_bbox_txt_path, 'r') as object_bbox_txt:
    for line in object_bbox_txt:
        x, y, w, h, conf, class_ = map(float, line.split())
        img = cv2.rectangle(img, (int(x), int(y)), (int(x+w), int(y+h)), color=(0, 0, 255), thickness=2)
    plt.imshow(img[:, :, ::-1])
```
