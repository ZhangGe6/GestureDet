```python
import os
import cv2
import matplotlib.pyplot as plt
```

```python
raw_data_dir = '../_LABELLED_SAMPLES'
bbox_dir = '../hand_bboxes'

seq_name = 'CHESS_LIVINGROOM_B_S'
frame_id = 'frame_0051'

img_path = os.path.join(raw_data_dir, seq_name, frame_id + '.jpg')
hand_bboxes_txt_path = os.path.join(bbox_dir, seq_name, frame_id + '_hand_bboxes.txt')
img = cv2.imread(img_path)
with open(hand_bboxes_txt_path, 'r') as f:
    for line in f:
        x, y, w, h = map(int, line.split())
        img = cv2.rectangle(img, (int(x), int(y)), (int(x+w), int(y+h)), color=(0, 0, 255), thickness=2)
plt.imshow(img[:, :, ::-1])
```

```python
raw_data_dir = '../_LABELLED_SAMPLES'
bbox_dir = '../objects_bboxes'

seq_name = 'CHESS_LIVINGROOM_B_S'
frame_id = 'frame_0051'

img_path = os.path.join(raw_data_dir, seq_name, frame_id + '.jpg')
hand_bboxes_txt_path = os.path.join(bbox_dir, seq_name, frame_id + '_objects_bboxes.txt')
img = cv2.imread(img_path)
with open(hand_bboxes_txt_path, 'r') as f:
    for line in f:
        print(line.split())
        x, y, w, h, conf, cls_ = map(float, line.split())
        img = cv2.rectangle(img, (int(x), int(y)), (int(x+w), int(y+h)), color=(0, 0, 255), thickness=2)
        img = cv2.putText(img, str(cls_), (int(x), int(y)), cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,255), 1)
plt.imshow(img[:, :, ::-1])
```
